Find all posts which are reacted to with a heart
Finad all posts which are reacted to with a heart. For such posts output all columns from facebook_posts table
Logic- find post reaction="heart"
Match those posts with facebook_posts return full post details

```sql
%sql
select post_id
from facebook_reactions
where reaction="heart";
```

```sql
%sql
select p.*
from facebook_posts p
inner join facebook_reactions r
on p.post_id=r.post_id
where r.reaction="heart";
```
