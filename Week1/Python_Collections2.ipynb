```python
# isalpha()- is a string method which is used to check if the string contain only letters
# returns boolean result
# True ✅ → if all characters are letters
# False ❌ → if there are numbers, spaces, or symbols
name="Bipina"
name.isalpha()
```

```python
#for- when looping through a dict, the return value are the values of the dict, but there are methods to return the keys as well 
```

```python
employee1={
    "id": 1,
    "name":"Bipina",
    "salary":20000,
      2:3
}
print(employee1)
```

```python
for x in employee1:
    print(employee1[x])
```

```python
for x in employee1.values():
    print(x)
```

```python
for x in employee1.keys():
    print(x)
```

```python
for x in employee1.keys():
    print(x)
```

```python
employee1={
    "id": 1,
    "name":"Bipina",
    "salary":20000,
      2:3
}
for key, value in employee1.items():
    print(key, value)
```

```python
for key in employee1.items():
    print(key)
```
