```python
# Zip function- it combines multiple collections into pairs, it groups element based on index position
coders=("Bipina", "Riya", "Aryan")
marks=(100, 20, 50)
# First zip function will return zip object, zip stops at the shortest list
coder_mark=zip(coders, marks)
list(coder_mark)
```

```python
for name, score in zip(coders, marks):
    print(name, score)
```

Difference between method and function
Methods → tied to an object, modify or act on it (e.g., list.append())
Functions → standalone, take object as input (e.g., min(list), zip())

```python
coders_mark=[('Bipina', 100), ('Riya', 20), ('Aryan', 50)]
coders, marks=zip(*coders_mark)
print(coders)
print(marks)
```

```python
this_dict1={
    "name": "Bipina",
    "course":"database",
    "year":2026,
    "year":2000
}
print(this_dict1)
```

```python
print(this_dict1.get("city")) #safest way to check if any item is present or not
```

```python
this_dict1.pop("city", "not found") #pop method removes key value pairs using the key
```

```python
# length of the dictionary
len(this_dict1)  #len function counts pairs(key+value)
```

```python
this_dict1.clear()  #clear removes everything from the dict but variable still exists.
print(this_dict1)
```
