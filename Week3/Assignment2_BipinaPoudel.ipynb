Question 2 - Find departments with no employees(joins)

```python
%sql
SELECT d.dept_name
FROM department d
LEFT JOIN employee e ON d.dept_id = e.dept_id
WHERE e.emp_id IS NULL;
```

Question 3 - Find employees without department(joins)

```python
%sql
SELECT e.*
FROM employee e
LEFT JOIN department d ON e.dept_id = d.dept_id
WHERE d.dept_id IS NULL;
```

Question 4 - Find second highest salary

```python
%sql
SELECT MAX(salary) AS second_highest
FROM employee
WHERE salary < (SELECT MAX(salary) FROM employee);
```

Question 5 - Count employees hired per year

```python
%sql
SELECT 
    YEAR(hire_date) AS hire_year,
    COUNT(*) AS emp_count
FROM employee
GROUP BY hire_year
ORDER BY hire_year;
```

Question 6 - Calculate running total of salary(Window Function)

```python
%sql
SELECT 
    emp_id,
    salary,
    SUM(salary) OVER (ORDER BY emp_id) AS running_total
FROM employee;
```
